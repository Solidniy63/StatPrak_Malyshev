```python
import pandas as pd
data = pd.read_csv('/content/sample_data/TableM.csv')
data['Project'] = data['Project'] * 2 #Делаем вес за проект более весомым чем за другие задания на 1 балл (обычные дз оцениваем в 1 балл)
data['Pandas'] = data['Pandas'] * 1.5  #Баллы за дз Pandas на 0.5 выше, чем за другие дз
data = data.fillna(-1) # Заполняем 0 пустые ячейки, так как студент решил не позориться кривым или неработающим кодом
results = data.iloc[:, 1:].sum(axis=1)
results = pd.concat([data['Student'], results], axis=1)
results = results.rename(columns={0: 'result'})

results
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(results['result'].values.reshape(-1, 1))
results['Cluster'] = kmeans.labels_
results['Cluster'].replace({0: 'автомат', 1: 'на зачёт', 2: 'пересдача'}, inplace=True)
results
```
